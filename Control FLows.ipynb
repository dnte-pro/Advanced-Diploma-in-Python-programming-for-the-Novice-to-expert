### iii) While Loops

```python
#Where while loop is preferred
for i in range(10):
    print(i)
```

```python
i = 0 
while i < 10 :
    print(i)
    i += 1
```

```python
my_list = [10,20,30,5,5,20,40]
res_add = 0
index = 0
while res_add < 70:
    res_add = res_add + my_list[index]
    index +=1
    print(res_add)
print("The sum (",res_add,") comes after adding the first",index,"elements")
```

## iv) Break Statements

```python
my_list = [1,2,3,"stop", 4.1]
for i in my_list:
    print(i)
    if i == "stop":
        print("The loop will stop")
        break
print("This is outside the loop")
```

Nested loop 

```python
my_list = [1,2,3,"stop", 4.1]
for i in my_list:
    print(i)
    for i in my_list:
        print (i)
        if i == "stop":
            print("The loop will stop")
            break
print("This is outside the loop")
```

```python
index = 0
while index < len(my_list):
    current_element = my_list[index]
    print(current_element)
    if current_element == "stop":
        print("The loop will stop")
        break
    index += 1
```

### Continue Statements

Used when you want to achieve some iterations

```python
my_list = [1, 2, 3, 'stop', 5, 'stop', 2]
for i in my_list:
    if i == 'stop':
        break
    print(i)
```

```python
my_list = [1, 2, 3, 'stop', 5, 'stop', 2]
for i in my_list:
    if i == 'stop':
        continue
    print(i)
```

 ### NOTE: 
 ##### The "continue" only skips the "stop"

### Zip function

```python
first_list = [1,2,"stop", 4.42]   # iterable
sec_list = [5,1,10,3]             # iterable
zipped_list = zip(first_list, sec_list)   #iterator
print(zipped_list)
```

```python
for i1, i2 in zipped_list:
    print(i1,i2)
```

```python
print(list(zipped_list))
```
