## Basic Condional Statements

```python
x =  10
y = 20 
if y > x:
    print(y)
```

```python
x = 10  
y = 20 
z = 30

if y > x:
    print(y)
elif z > y:
    print(z)
else:
    print(x)

print("This is printed either way because it is outside the conditional statement.")
```

```python
if y > x:
    print("This is the first block!")
    if z > y:
        print(z)
    elif y > x: #afte a " True" has been found the code does not execute thereafter
        print(y)
    print("This is the first block also")
print("This is block zero")
```
