# Arithmetic Operators

#### Arithmetic operators

```python
first_num = 5
sec_num = 3.0

#Add
result = first_num + sec_num

print(result)
print(type(result))
```

```python
# substract 
result_subtract = first_num - sec_num
print(result_subtract)
print(type(result_subtract))
```

```python
result_div = first_num / sec_num
print(result_div)
print(type(result_div))
```

```python
result_mul = first_num * sec_num
print(result_mul)
print(type(result_mul))
```

```python
result_rem = first_num % sec_num
print(result_rem)
```

```python
result_divfloor = first_num // sec_num
print(result_divfloor)
```
