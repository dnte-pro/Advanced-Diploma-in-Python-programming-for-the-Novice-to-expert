## Handling Exceptions

```python
def div(x,y):
    try:
        result = x/y
    except:                        #the except helps handle the exception errors
        result = "Warning"
    return result
```

```python
print(div(1,2))
print(div(1,"a"))
print("Hello")
```

```python
def div(x,y):
    try:
        assert x==y 
        result = x/y
    except:
        result = "Can't divide by zero"
    return result
print(div(5,2))
print(div(5,0))
print("Hello World")
```

#### Handling multiple exceptions

```python
def div(x,y):
    try:
        result = x/y
        assert x >= y
    except ZeroDivisionError:
        result = "Can't Divide by zero"
    except AssertionError:
        print("Error, z is less than y")
    return result

print(div(6,3))
print(div(4,8))
print(div(5,0))
print("Hello World")
```

Now using the Except and else

```python
def div(x,y):
    
    try:
        result = x/y
    except ZeroDivisionError:
        result = "Can't divide by zero"
    else:
        print("Else")
    return result

print(div(6,3))
print(div(4,8))
print(div(5,0))
print("Hello World")
```

```python
def bool_return():
    try:
        return True
    finally:
        return False
bool_return()
```

```python
def div(x,y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Division by zero")
    else:
        print("Result is", result)
    finally:
        print("Executing finally")

div(2,1)
div(2,0)
div("2","1")
```

## Map Function

```python
def add_five(my_list):
    output_list =[]
    for element in my_list:
        element = element + 5
        output_list.append(element)
    return output_list
```

```python
my_list = [5,1,20]
add_five(my_list)
```

```python
def add_five_map(element):
    return element + 5
output = map(add_five_map, my_list)
```

```python
type(output)
```

## Filter Function

```python
def pass_grade_normal(list_grades):
    pass_grade_list = []
    for current_grade in list_grades:
        if current_grade > 50:
            pass_grade_list.append(current_grade)
    return pass_grade_list
```

```python
list_grade = [60, 55, 100, 20, 34, 70]
```

```python
pass_grade_normal(list_grade)
```

Using the filter function

```python
def pass_grade_filter(grade):
    return grade> 50
```

```python
passed_grade_filter = list(filter(pass_grade_filter, list_grade))
```
