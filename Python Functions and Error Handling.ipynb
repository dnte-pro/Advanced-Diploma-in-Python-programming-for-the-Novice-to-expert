## Handling Exceptions

```python
def div(x,y):
    try:
        result = x/y
    except:                        #the except helps handle the exception errors
        result = "Warning"
    return result
```

```python
print(div(1,2))
print(div(1,"a"))
print("Hello")
```

```python
def div(x,y):
    try:
        assert x==y 
        result = x/y
    except:
        result = "Can't divide by zero"
    return result
print(div(5,2))
print(div(5,0))
print("Hello World")
```

#### Handling multiple exceptions

```python
def div(x,y):
    try:
        result = x/y
        assert x >= y
    except ZeroDivisionError:
        result = "Can't Divide by zero"
    except AssertionError:
        print("Error, z is less than y")
    return result

print(div(6,3))
print(div(4,8))
print(div(5,0))
print("Hello World")
```

Now using the Except and else

```python
def div(x,y):
    
    try:
        result = x/y
    except ZeroDivisionError:
        result = "Can't divide by zero"
    else:
        print("Else")
    return result

print(div(6,3))
print(div(4,8))
print(div(5,0))
print("Hello World")
```

```python
def bool_return():
    try:
        return True
    finally:
        return False
bool_return()
```

```python
def div(x,y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Division by zero")
    else:
        print("Result is", result)
    finally:
        print("Executing finally")

div(2,1)
div(2,0)
div("2","1")
```
